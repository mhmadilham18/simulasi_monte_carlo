```python
headers = {
    "Authorization": f"Bearer {TOKEN_OPERATOR}",
    "Content-Type": "application/json",
    "Accept": "application/json"
}
```

```python
def generate_password(length=8):
    """Generate password random kombinasi huruf dan angka."""
    characters = string.ascii_letters + string.digits
    return ''.join(random.choice(characters) for i in range(length))
```

```python
def generate_username(nama_lengkap):
    nama_depan = str(nama_lengkap).split()[0].lower()
    nama_depan = ''.join(e for e in nama_depan if e.isalnum())
    angka_random = random.randint(100, 999)
    return f"{nama_depan}_{angka_random}"
```

```python
def main():
    print(f"Membaca data dari {FILE_CSV_INPUT}...")
    try:
        df = pd.read_csv(FILE_CSV_INPUT, dtype={'nisn': str})
    except FileNotFoundError:
        print(f"Error: File {FILE_CSV_INPUT} tidak ditemukan!")
        return

    hasil_generate = []

    print(f"Ditemukan {len(df)} data siswa. Memulai pembuatan akun...")
    
    for index, row in df.iterrows():
        nama_siswa = row['nama_siswa']
        nisn = str(row['nisn']).strip()
        
        email_siswa = f"{nisn}@student.sch.id"
        username_siswa = generate_username(nama_siswa)
        password_siswa = generate_password()
        
        payload = {
            "email": email_siswa,
            "username": username_siswa,
            "full_name": nama_siswa.title(),  
            "password": password_siswa,
            "identity_number": nisn
        }
        
        print(f"[{index + 1}/{len(df)}] Memproses: {nama_siswa} (NISN: {nisn})...", end=" ")
        
        try:
            response = requests.post(ENDPOINT, json=payload, headers=headers)
            
            if response.status_code in [200, 201]:
                print("SUKSES")
                hasil_generate.append({
                    "Nama Siswa": nama_siswa.title(),
                    "NISN": nisn,
                    "Username": username_siswa,
                    "Email": email_siswa,
                    "Password": password_siswa,
                    "Status": "Sukses"
                })
            else:
                print(f"GAGAL (Status: {response.status_code}) - {response.text}")
                hasil_generate.append({
                    "Nama Siswa": nama_siswa.title(),
                    "NISN": nisn,
                    "Username": "-",
                    "Email": "-",
                    "Password": "-",
                    "Status": f"Gagal API: {response.status_code}"
                })
                
        except Exception as e:
            print(f"ERROR JARINGAN: {e}")
            hasil_generate.append({
                "Nama Siswa": nama_siswa.title(),
                "NISN": nisn,
                "Username": "-",
                "Email": "-",
                "Password": "-",
                "Status": f"Error: {str(e)}"
            })
            
        time.sleep(0.5) 

    print("\nProses selesai. Menyimpan hasil ke Excel...")
    df_hasil = pd.DataFrame(hasil_generate)
    
    try:
        df_hasil.to_excel(FILE_EXCEL_OUTPUT, index=False)
        print(f"Berhasil! File Excel tersimpan dengan nama: {FILE_EXCEL_OUTPUT}")
    except Exception as e:
        print(f"Gagal menyimpan file Excel: {e}")
```

```python
if __name__ == "__main__":
    main()
```
